```python
f = open('input5.txt')
data = f.readlines()
# data
```

```python
print(data[0])
print(data[0].replace('F','0'))
print(int(data[0].replace('L','0').replace('R','1')[7:],2))
```

```python
max_num = 0
seats = {}
for bp in data:
    row = int(bp.replace('F','0').replace('B','1')[:7],2)
    col = int(bp.replace('L','0').replace('R','1')[7:],2)
    seat_id = row*8+col
    if seat_id > max_num:
        max_num = seat_id
    if row not in seats:
        seats[row] = []
    seats[row].append(col)
#     print(row, col, seat_id, max_num)
# seats
for i in seats:
    if len(seats[i]) != 8:
        print(i,seats[i])
# max_num
```
